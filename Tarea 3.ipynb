```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
hbar = 1.0  
mass = 1.0
```

```python
x_min = -5.0  
x_max = 5.0   
N = 1000      
dx = (x_max - x_min) / N  
```

```python
def potential(x):
    return 0.5 * x**2
x = np.linspace(x_min, x_max, N)
psi = np.zeros(N)
V = potential(x)
```

```python
h = x[1] - x[0]
psi = np.zeros(N)
psi[0] = 0.0
psi[1] = 1e-5

def numerov_method(x,v,E):
    h=x[1]-x[0]  
    
    for i in range(2, N):
        rn=2*(v[i]-E)
        rminus2=2*(v[i-2]-E)
        rminus1=2*(v[i-1]-E)
        gamma=1-rn*h**2/12
        epsilon=1+5*h**2*rminus1/12
        alpha=h**2*rminus2/12-1
        psi[i] = (2*epsilon*psi[i-1]+alpha*psi[i-2])/gamma

    return psi
def valores_propios(psi,E_seed,n,h,x,v):
    #primero calcular psi para E_seed
    #n es el número de niveles de energía

    eigen_energy=np.zeros((n,1))
    for i in range(int(n)):
        
        while psi(x,v,E_seed)[-1]*psi(x,v,E_seed+h)[-1]>0:
            E_seed=E_seed+h
        else:
            eigen_energy[i,0]=E_seed
            E_seed=E_seed+h
    return eigen_energy
```

```python
E = 0.5
psi_values = numerov_method(x,V, E)
plt.plot(x,psi_values)
```

```python
energies=valores_propios(numerov_method,0.3,6,0.01,x,V)

energies

```

```python
for i in range(len(energies)):
    psi=numerov_method(x,V,energies[i])
    max_value = np.max(np.abs(psi))
    psi_normalized = psi / max_value
 
    plt.plot(x,psi_normalized,label="E= "+str(energies[i]))
plt.xlabel("x")
plt.ylabel( r'$\Psi$'+"(x)")
plt.legend()
```

```python
plt.plot(x, V, label='Potential Energy')
plt.plot(x,psi_normalized)
plt.xlabel('Position (x)')
plt.ylabel('Wavefunction / Potential Energy')
plt.legend()
plt.title('Numerical Solution of Schrödinger Equation')
plt.show()
```
